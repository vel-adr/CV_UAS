```python
import os
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from skimage.io import imread
import matplotlib.image as mpimg
from skimage.transform import resize
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# 1B
dataset_path = "../input/animal-image-datasetdog-cat-and-panda/animals/animals"
labels = []

for i in os.listdir(dataset_path):
    labels.append(i)
    
print("Jumlah kelas:", len(labels))
print("\nKelas yang ada:")
for item in labels:
    print(item)
```

```python
imgs =[]

example_imgs_path = "../input/animal-image-datasetdog-cat-and-panda/images"
for i in os.listdir(example_imgs_path):
    img_path = example_imgs_path + '/' + i
    imgs.append([i, mpimg.imread(img_path)])
  
fig = plt.figure(figsize=(10,10))
i=1
for item in imgs:
    fig.add_subplot(3,3,i)
    plt.imshow(item[1])
    plt.title(item[0].replace(".jpg",""))
    i+=1
plt.show()
```

```python
import splitfolders

output_path = "./splitted_dataset"
splitfolders.ratio(dataset_path, output=output_path, seed=1337, ratio=(0.8, 0.2))
```

```python
train_path = output_path + '/train'
val_path = output_path + '/val'

train_imgs = []
val_imgs = []
for lbl in os.listdir(train_path):
    lbl_path = train_path+'/'+lbl
    for img in os.listdir(lbl_path):
        train_imgs.append(img)
        
for lbl in os.listdir(val_path):
    lbl_path = val_path+'/'+lbl
    for img in os.listdir(lbl_path):
        val_imgs.append(img)
        
print('Jumlah image pada training dataset: ' + str(len(train_imgs)))
print('Jumlah image pada validation dataset: ' + str(len(val_imgs)))
```

```python
input_data_arr = []
target_arr = []

# Memuat semua image dan label
for i in labels:
    path = os.path.join(dataset_path, i)
    for img in os.listdir(path):
        img_array = imread(os.path.join(path,img))
        img_resized=resize(img_array,(150,150,3))
        input_data_arr.append(img_resized.flatten())
        target_arr.append(labels.index(i))

input_data = np.array(input_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(input_data)
df['Target']=target
x=df.iloc[:,:-1]
y=df.iloc[:,-1]

# Memisahkan dataset menjadi data training dan data test dengan ratio 80% - 20%
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
# 1C
# SVM
# Inisiasi model SVM
svm = SVC(C=5)

# Memulai proses training
svm.fit(x_train,y_train)
```

```python
# Menghitung dan menampilkan nilai akurasi dari model SVM
acc = svm.score(x_test, y_test)
print("SVM Model Accuracy: " + str(acc*100) + '%')
```

```python
#KNN
# Inisiasi model KNN
model = KNeighborsClassifier(n_neighbors=7)

# Memulai proses training
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)

print("kNN Model Accuracy:" + str(acc * 100) + '%')
```
